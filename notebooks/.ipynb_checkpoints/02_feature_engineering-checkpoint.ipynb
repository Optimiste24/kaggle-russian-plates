# Feature Engineering - Russian Plates
### **Objectif** : Créer des variables avancées pour la modélisation

````python

import pandas as pd
import numpy as np
from pathlib import Path

# Chargement des données pré-nettoyées (output de l'EDA)
data_path = Path("C:/Users/Optimiste/Videos/Concours/Prédiction_prix_plaques_russes/data")
train = pd.read_csv(data_path / "train_eda.csv") 
test = pd.read_csv(data_path / "test_eda.csv")
````

---
### 1. Plaques Premium (Version Optimisée)

````python
## 2.1 Flags Spéciaux
# Liste étendue des combinaisons prisées
premium_letters = ['AA', 'KK', 'XX', 'TT', 'MM', 'OO', 'AK', 'KC', 'CC']

test['premium_plate'] = (
    (test['is_gov_plate'] == 1) |
    (test['digits'].isin([111, 222, 333, 444, 555, 666, 777, 888, 999])) |
    (test['letters'].str.contains('|'.join(premium_letters)))
).astype(int)

# Vérification
print("% de plaques premium :", test['premium_plate'].mean().round(3))
````

````python
## 2.1 Flags Spéciaux
# Liste étendue des combinaisons prisées
premium_letters = ['AA', 'KK', 'XX', 'TT', 'MM', 'OO', 'AK', 'KC', 'CC']

train['premium_plate'] = (
    (train['is_gov_plate'] == 1) |
    (train['digits'].isin([111, 222, 333, 444, 555, 666, 777, 888, 999])) |
    (train['letters'].str.contains('|'.join(premium_letters)))
).astype(int)

# Vérification
print("% de plaques premium :", train['premium_plate'].mean().round(3))
````

---
### 2. Rareté des Combinaisons

````python
import matplotlib.pyplot as plt
import seaborn as sns

##  Rareté des Lettres
# Fréquence relative des combinaisons
letter_freq = train['letters'].value_counts(normalize=True)
train['letters_rarity'] = train['letters'].map(letter_freq)

# Classification en tiers
train['letters_rarity_class'] = pd.qcut(
    train['letters_rarity'],
    q=3,
    labels=['common', 'medium', 'rare']
)

# Visualisation
pd.crosstab(train['letters_rarity_class'], train['premium_plate']).plot.bar()
plt.title('Rareté vs Statut Premium')
plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

##  Rareté des Lettres
# Fréquence relative des combinaisons
letter_freq = test['letters'].value_counts(normalize=True)
test['letters_rarity'] = test['letters'].map(letter_freq)

# Classification en tiers
test['letters_rarity_class'] = pd.qcut(
    test['letters_rarity'],
    q=3,
    labels=['common', 'medium', 'rare']
)

# Visualisation
pd.crosstab(test['letters_rarity_class'], test['premium_plate']).plot.bar()
plt.title('Rareté vs Statut Premium')
plt.show()
````

---
### 3. Extraction des Années

````python
# Assure-toi d'abord que la colonne 'date' est bien au format datetime
train['date'] = pd.to_datetime(train['date'])
test['date'] = pd.to_datetime(test['date'])

# Version robuste pour calculer l'âge des plaques
def calculate_plate_age(df):
    df = df.copy()
    # Trouve la première apparition de chaque plaque
    first_listings = df.groupby('plate')['date'].min().rename('first_listing_date')
    
    # Fusionne avec le dataframe principal
    df = df.merge(first_listings, how='left', on='plate')
    
    # Calcule l'âge en jours
    df['plate_age_days'] = (df['date'] - df['first_listing_date']).dt.days
    
    return df.drop(columns=['first_listing_date'])

train = calculate_plate_age(train)
test = calculate_plate_age(test)
````

````python
#  Features Temporelles Avancées

# 1. Périodicité
train['listing_dayofweek'] = train['date'].dt.dayofweek  # Lundi=0, Dimanche=6
train['listing_week'] = train['date'].dt.isocalendar().week
train['listing_quarter'] = train['date'].dt.quarter

# 2. Dernière apparition (pour rareté dynamique)
last_listings = train.groupby('plate')['date'].max().rename('last_listing_date')
train = train.merge(last_listings, how='left', on='plate')
train['days_since_last_seen'] = (train['date'] - train['last_listing_date']).dt.days

# 3. Saisonnalité (ex: prix plus hauts en été)
seasons = {
    1: 'winter', 2: 'winter', 3: 'spring', 
    4: 'spring', 5: 'spring', 6: 'summer',
    7: 'summer', 8: 'summer', 9: 'fall',
    10: 'fall', 11: 'fall', 12: 'winter'
}
train['season'] = train['date'].dt.month.map(seasons)

# Vérification
print("\nExemple de features temporelles :")
print(train[['plate', 'date', 'plate_age_days', 'season']].head())
````

````python
#  Features Temporelles Avancées

# 1. Périodicité
test['listing_dayofweek'] = test['date'].dt.dayofweek  # Lundi=0, Dimanche=6
test['listing_week'] = test['date'].dt.isocalendar().week
test['listing_quarter'] = test['date'].dt.quarter

# 2. Dernière apparition (pour rareté dynamique)
last_listings = test.groupby('plate')['date'].max().rename('last_listing_date')
test = test.merge(last_listings, how='left', on='plate')
test['days_since_last_seen'] = (test['date'] - test['last_listing_date']).dt.days

# 3. Saisonnalité (ex: prix plus hauts en été)
seasons = {
    1: 'winter', 2: 'winter', 3: 'spring', 
    4: 'spring', 5: 'spring', 6: 'summer',
    7: 'summer', 8: 'summer', 9: 'fall',
    10: 'fall', 11: 'fall', 12: 'winter'
}
test['season'] = test['date'].dt.month.map(seasons)

# Vérification
print("\nExemple de features temporelles :")
print(test[['plate', 'date', 'plate_age_days', 'season']].head())
````

````python
# %% [markdown]
## Vérification des Features Temporelles
# %%
plt.figure(figsize=(15, 5))

# 1. Évolution des prix par âge de plaque
plt.subplot(131)
sns.scatterplot(x='plate_age_days', y='price', data=train.sample(1000), alpha=0.5)
plt.title('Prix vs Âge de la plaque')

# 2. Prix moyen par saison
plt.subplot(132)
sns.boxplot(x='season', y='price', data=train, showfliers=False)
plt.title('Distribution des prix par saison')

# 3. Volume de transactions par jour de semaine
plt.subplot(133)
train['listing_dayofweek'].value_counts().sort_index().plot.bar()
plt.title('Transactions par jour de semaine')

plt.tight_layout()
plt.show()
````

# %%
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Calcul séquentiel pour éviter les matrices géantes
gov_scaled = MinMaxScaler().fit_transform(train[['gov_importance_level']]).ravel()
digits_scaled = train['digits'].values / 999
rarity_scaled = 1 - train['letters_rarity'].values

# Combinaison pondérée
train['prestige_score'] = 0.4 * gov_scaled + 0.3 * digits_scaled + 0.3 * rarity_scaled

# Version alternative encore plus légère (sans MinMaxScaler)
train['prestige_score_alt'] = (
    0.4 * (train['gov_importance_level'] / train['gov_importance_level'].max()) +
    0.3 * (train['digits'] / 999) +
    0.3 * (1 - train['letters_rarity'])




# %% [markdown]
### 3.2 Optimisation Mémoire
# %%
# Libération de mémoire
import gc
del gov_scaled, digits_scaled, rarity_scaled
gc.collect()

# Traitement par chunks si nécessaire
def process_in_chunks(df, chunk_size=10000):
    results = []
    for i in range(0, len(df), chunk_size):
        chunk = df.iloc[i:i+chunk_size]
        res = 0.4*(chunk['gov_importance_level']/10) + 0.3*(chunk['digits']/999) + 0.3*(1-chunk['letters_rarity'])
        results.append(res)
    return pd.concat(results)

train['prestige_score_chunked'] = process_in_chunks(train)

### 5. Sauvegarde professionnelle des résultats

````python
# %% [markdown]
### 4. Sauvegarde des Résultats en CSV
# %%
from pathlib import Path
import pandas as pd
import numpy as np

# Dossier de sortie
output_dir = Path("C:/Users/Optimiste/Videos/Concours/Prédiction_prix_plaques_russes/output")
output_dir.mkdir(parents=True, exist_ok=True)

# Sélection des colonnes essentielles (adaptée pour CSV)
essential_features = [
    'id',
    'plate',
    'date',
    'price',
    # Features de base
    'digits', 
    'letters',
    'region_code',
    'region_name',
    # Flags importants
    'is_gov_plate',
    'premium_plate',
    # Features temporelles
    'plate_age_days',
    'season'
]

# Conversion des types pour réduire la taille
dtypes = {
    'id': 'int32',
    'digits': 'int16',
    'is_gov_plate': 'int8',
    'premium_plate': 'int8',
    'plate_age_days': 'int16'
}

# Application des types et sélection des colonnes
train_export = train[essential_features].astype(dtypes)
test_export = test[essential_features].astype(dtypes)

# Sauvegarde avec compression (gzip réduit la taille de 70%)
train_export.to_csv(
    output_dir / "train_processed.csv.gz",
    index=False,
    compression='gzip'
)

test_export.to_csv(
    output_dir / "test_processed.csv.gz",
    index=False,
    compression='gzip'
)

# Documentation des features
feature_docs = {
    'premium_plate': "Combinaison de plaques gouvernementales/chiffres magiques/lettres rares (0/1)",
    'plate_age_days': "Nombre de jours depuis la première apparition de la plaque",
    'season': "Saisonnalité: winter/spring/summer/fall"
}

pd.Series(feature_docs).to_csv(output_dir / "features_description.csv")

print(f"""
✅ Fichiers sauvegardés avec succès dans {output_dir} :
- train_processed.csv.gz ({len(train_export):,} lignes)
- test_processed.csv.gz ({len(test_export):,} lignes)
- features_description.csv
""")
````
