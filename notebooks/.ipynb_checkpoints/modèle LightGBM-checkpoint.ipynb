```python
#  Soumission Kaggle - LightGBM

import pandas as pd
import numpy as np
from pathlib import Path
from lightgbm import LGBMRegressor
from lightgbm import early_stopping
import joblib

# Chargement des données
data_path = Path("C:/Users/Optimiste/Videos/Concours/Prédiction_prix_plaques_russes/data")

train = pd.read_csv(data_path / "train_preprocessed_v2.csv")
test = pd.read_csv(data_path / "test_preprocessed_v2.csv")
preprocessors = joblib.load(data_path / "preprocessors.joblib")

# Features sélectionnées
features = [
    'digits', 'region_encoded', 'is_gov_plate',
    'premium_plate', 'plate_age_days', 'premium_region'
]

X_train = train[features]
y_train = train['log_price']
X_test = test[features]

# Hyperparamètres optimisés
lgbm_params = {
    'n_estimators': 850,           # Augmenté pour meilleure performance
    'learning_rate': 0.05,         # Taux d'apprentissage réduit
    'num_leaves': 31,              # Contrôle la complexité
    'force_row_wise': True,
    'random_state': 42,
    'verbosity': -1,
    'metric': 'mape'
}

# Entraînement final
model = LGBMRegressor(**lgbm_params)
model.fit(
    X_train, y_train,
    eval_set=[(X_train, y_train)],
    eval_metric='mape',
    callbacks=[early_stopping(50)]  # Arrêt précoce si pas d'amélioration
)
```

**Validation Croisée Rigoureuse**

```python
from sklearn.model_selection import TimeSeriesSplit, cross_val_score
from sklearn.metrics import make_scorer
import numpy as np

def lgbm_smape(y_true, y_pred):
    return 100 * np.mean(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)))

cv_scores = cross_val_score(
    model,
    X_train,
    y_train,
    cv=TimeSeriesSplit(n_splits=5),
    scoring=make_scorer(lgbm_smape, greater_is_better=False)
)

print(f"SMAPE moyen en CV: {-cv_scores.mean():.4f} (±{cv_scores.std():.4f})")
```

**Génération des Prédiction**

```python
# %% [markdown]
## 3. Prédictions pour Soumission
# %%
# Prédiction sur le test set
test_preds_log = model.predict(X_test)

# Conversion inverse (log → prix réels)
test_preds = np.expm1(test_preds_log)

# Post-traitement
test_preds = np.clip(test_preds, 0, None)  # Évite les prix négatifs

# Formatage pour Kaggle
submission = pd.DataFrame({
    'ID': test['id'],
    'Price': test_preds.round(2)  # 2 décimales
})
```

**Optimisations complémentaires**

```python
# %% [markdown]
## 4. Optimisations Finales
# %%
# Ajustement pour les plaques premium
premium_mask = test['premium_plate'] == 1
submission.loc[premium_mask, 'Price'] *= 1.15  # +15% pour plaques premium

# Correction des outliers
q99 = np.quantile(submission['Price'], 0.99)
submission['Price'] = np.where(
    submission['Price'] > q99,
    q99,
    submission['Price']
)

# Vérification finale
print("\nStatistiques des prédictions :")
print(submission['Price'].describe())
```

**Vérification Finale des Prédictions**

```python
#  Vérification des Distributions
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.hist(np.expm1(y_train), bins=50, alpha=0.7, label='Train')
plt.hist(submission['Price'], bins=50, alpha=0.7, label='Test Pred')
plt.xlabel('Price (RUB)')
plt.ylabel('Frequency')
plt.legend()
plt.title('Distribution Comparaison')

plt.subplot(122)
plt.scatter(np.expm1(y_train), np.expm1(model.predict(X_train)), alpha=0.3)
plt.plot([0,2e6], [0,2e6], 'r--')
plt.xlabel('True Price')
plt.ylabel('Predicted Price')
plt.title('Prediction Accuracy')
plt.tight_layout()
plt.show()
```

**Sauvegarde des Artefacts**

```python
#  Sauvegarde des Fichiers
import joblib
from datetime import datetime

# Dossier de sauvegarde
model_dir = Path("C:/Users/Optimiste/Videos/Concours/Prédiction_prix_plaques_russes/model")
model_dir.mkdir(exist_ok=True)

# Nom unique basé sur la date
timestamp = datetime.now().strftime("%Y%m%d_%H%M")
model_name = f"lgbm_model_{timestamp}"

# Sauvegarde
joblib.dump(model, model_dir / f"{model_name}.joblib")
submission.to_csv(model_dir / f"submission_{model_name}.csv", index=False)

# Fichier de métadonnées
with open(model_dir / f"metadata_{model_name}.txt", "w") as f:
    f.write(f"""Modèle LightGBM - Soumission Kaggle
Date: {timestamp}
SMAPE CV: 5.5513 (±0.0768)
Features utilisées: {features}
Hyperparamètres: {lgbm_params}
Statistiques prédictions:
{submission['Price'].describe()}
""")
```
